```python
filtered_taco = X['menus.name'].str.contains('Taco', flags=re.IGNORECASE, regex=True)
filtered_burrito = X['menus.name'].str.contains('Burrito', flags=re.IGNORECASE, regex=True)
X['Taco'] = filtered_taco
X['Burrito'] = filtered_burrito
```

```python
print(filtered_taco.value_counts())
print(filtered_burrito.value_counts())
```

```python
X['menu_type'] = 'Unknown'
X['menu_type'] = np.where(X['Taco'] == True, 'Taco', X['menu_type'])
X['menu_type'] = np.where(X['Burrito'] == True, 'Burrito', X['menu_type'])
```

```python
X.drop(['Taco', 'Burrito'], axis=1)
```

```python
X.to_csv('tacos-burritos-vs-city.csv')
```
